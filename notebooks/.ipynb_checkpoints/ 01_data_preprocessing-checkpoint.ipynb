# **Data Preprocessing & Cleaning**

```python
import pandas as pd
import numpy as np
```

## 1. Load Dataset

```python
file_path = "../data/processed.cleveland.data"

columns = [
    "age","sex","cp","trestbps","chol","fbs","restecg",
    "thalach","exang","oldpeak","slope","ca","thal","target"
]

df = pd.read_csv(file_path, names=columns)
print(df.head())
```

```python
df.shape
```

```python
print(df.info())
```

## 2. Missing Values

```python
# Replace "?" with NaN
df.replace("?", np.nan, inplace=True)

# Convert numeric columns to proper dtypes
df = df.apply(pd.to_numeric, errors="ignore")

# Check missing values again
print(df.isna().sum())
```

```python
print(df.info())
```

```python
# Categorical columns 
cat_cols = ["sex","cp","fbs","restecg","exang","slope","ca","thal"]

for col in cat_cols:
    print(f"{col}: {df[col].unique()}")
```

```python
# Check missing counts first
print(df[["ca", "thal"]].isna().sum())

# Impute with mode (most frequent value)
df["ca"] = df["ca"].fillna(df["ca"].mode()[0])
df["thal"] = df["thal"].fillna(df["thal"].mode()[0])

# Verify again
print(df[["ca", "thal"]].isna().sum())
```

```python
# Categorical columns 
cat_cols = ["sex","cp","fbs","restecg","exang","slope","ca","thal"]

for col in cat_cols:
    print(f"{col}: {df[col].unique()}")
```

```python
print(df.info())
```

## 3. Encoding
In your dataset:

- Keep as numeric (binary): sex, fbs, exang â†’ already 0/1

- One-hot encode (multi-category): cp, restecg, slope, ca, thal

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
multi_cat = ["cp", "restecg", "slope", "ca", "thal"]
encoder = OneHotEncoder(drop="first", sparse_output=False)
```

```python
# Fit-transform categorical data
encoded_array = encoder.fit_transform(df[multi_cat])

# Get new column names
encoded_cols = encoder.get_feature_names_out(multi_cat)

# Create DataFrame with encoded columns
encoded_df = pd.DataFrame(encoded_array, columns=encoded_cols,index=df.index)
```

```python
# Drop original categorical columns and join encoded ones
df_with_encoded = df.drop(columns = multi_cat).join(encoded_df)

print(df_with_encoded.head())
print(df_with_encoded.shape)
```

## 4. Standardize

```python
from sklearn.preprocessing import StandardScaler
```

```python
# Continuous columns to scale 
num_cols = ["age", "trestbps", "chol", "thalach", "oldpeak"]

scaler = StandardScaler()

# Scale only numeric columns, keep DataFrame intact
df_with_encoded[num_cols] = scaler.fit_transform(df_with_encoded[num_cols])

print(df_with_encoded.head())
```

## 5. EDA

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Histograms for numeric columns
df_with_encoded.hist(bins=30, figsize=(15, 12), edgecolor="black")
plt.suptitle("Histograms of Numeric Features", fontsize=16)
plt.show()
```

```python
# Compute correlation matrix
corr = df_with_encoded.corr()

plt.figure(figsize=(16, 10))
sns.heatmap(corr, annot=True, cmap="RdBu_r", center=0)
plt.title("Correlation Heatmap", fontsize=16)
plt.show()
```

```python
threshold = 0.4
strong_corr = corr.unstack().sort_values(ascending=False)
strong_corr = strong_corr[strong_corr.abs() >= threshold]
print(strong_corr)
```

```python
# Correlation with target only (exclude self-correlation)
target_corr = corr["target"].drop("target").sort_values(key=abs, ascending=False)

# Display top 10 features
print("Top correlations with target:\n", target_corr.head(10))

# Plot
plt.figure(figsize=(8,6))
target_corr.head(10).sort_values().plot(kind="barh", color="steelblue")
plt.title("Top 10 Correlations with Target", fontsize=14)
plt.xlabel("Correlation coefficient")
plt.ylabel("Features")
plt.show()
```

```python
# Boxplots for continuous variables
plt.figure(figsize=(15, 8))
for i, col in enumerate(num_cols, 1):
    plt.subplot(2, 3, i)
    sns.boxplot(y=df_with_encoded[col])
    plt.title(f"Boxplot of {col}")
plt.tight_layout()
plt.show()
```

```python
# Copy to avoid modifying original
df_out = df_with_encoded.copy()

# IQR-based filtering
for col in num_cols:  # continuous features
    Q1 = df_out[col].quantile(0.25)
    Q3 = df_out[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    df_out = df_out[(df_out[col] >= lower) & (df_out[col] <= upper)]

print("Original shape:", df_with_encoded.shape)
print("After outlier removal:", df_out.shape)
```

```python
# Boxplots for continuous variables
plt.figure(figsize=(15, 8))
for i, col in enumerate(num_cols, 1):
    plt.subplot(2, 3, i)
    sns.boxplot(y=df_out[col])
    plt.title(f"Boxplot of {col}")
plt.tight_layout()
plt.show()
```

```python
df_out.to_csv("../data/heart_cleaned.csv", index=False)
```
