# **Dimensionality Reduction - PCA** 
## 1. Apply PCA 

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/heart_cleaned.csv")
```

```python
from sklearn.decomposition import PCA
```

```python
X = df.drop(columns = ["target"])
y = df["target"]

# Apply PCA (keep all components first to analyze variance)
pca = PCA()
X_pca = pca.fit_transform(X)
```

Rule of thumb: Keep the smallest number of PCs that explain ~85–95% of the variance (depending on balance between dimensionality and info loss).

In your case:

- 10 PCs → 88% (pretty solid).

- 12 PCs → 93% (very good).

Beyond that, variance gains are small.

```python
explained_var = pca.explained_variance_ratio_
cumulative_var = np.cumsum(explained_var)

print("Explained variance per component:\n", explained_var)
print("Cumulative variance:\n", cumulative_var)
```

## 2. Optimal # of Components 

Instead of using all 20 features, you could reduce to ~**10–12** PCs without losing much information.

```python
# choose number of PCs to explain ~90% variance
n_components = np.argmax(cumulative_var >= 0.90) + 1
print("Optimal number of components (90% variance):", n_components)
```

```python
# Refit PCA using only the chosen number of components
pca_final = PCA(n_components=n_components)
X_pca_final = pca_final.fit_transform(df.drop(columns=["target"]))

# Build final PCA dataset
df_pca = pd.DataFrame(X_pca_final, columns=[f"PC{i+1}" for i in range(n_components)])
df_pca["target"] = df["target"].values
```

## 3. Visualize PCA 

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. Scree plot: variance per component + cumulative
plt.figure(figsize=(10,6))
plt.bar(range(1, len(explained_var)+1), explained_var, alpha=0.6, label="Individual explained variance")
plt.plot(range(1, len(cumulative_var)+1), cumulative_var, marker='o', color='red', label="Cumulative explained variance")
plt.xlabel("Principal Component")
plt.ylabel("Explained Variance Ratio")
plt.title("PCA Explained Variance")
plt.legend()
plt.show()
```

```python
# 2. Scatter plot of first 2 PCs colored by target
plt.figure(figsize=(8,6))
sns.scatterplot(
    x=X_pca[:,0], y=X_pca[:,1],
    hue=df["target"],
    palette="Set1", alpha=0.7
)
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("PCA - First Two Principal Components")
plt.legend(title="Target")
plt.show()
```

```python
df_pca.to_csv("../data/heart_pca.csv", index=False)
```
